```python
import os
import numpy as np
import numpy.ma as ma
import astropy.io.fits as fits
import matplotlib.pyplot as plt
from matplotlib import rc
from scipy import integrate
from scipy.stats import norm
from astropy.io import ascii
from astropy.table import Table
from scipy.optimize import minimize


np.random.seed(10)
rv = np.linspace(-2000, 2000, 1000)
y = np.hstack((norm.rvs(loc=0, scale=300, size=700), norm.rvs(loc=1200, scale=100, size=300)))
nums, v_edge = np.histogram(y, bins=np.arange(-2000, 2200, 200))


plt.figure()
plt.hist(y, bins=np.arange(-2000, 2200, 200))
A = np.sum(nums) * 200
plt.plot(rv, A * (0.7 * norm.pdf(rv, 0, 300) + 0.3 * norm.pdf(rv, 1200, 100)), '-b', lw=1, alpha=1)

plt.figure()
plt.hist(y, density=True, bins=np.arange(-2000, 2200, 200))
plt.plot(rv, (0.7 * norm.pdf(rv, 0, 300) + 0.3 * norm.pdf(rv, 1200, 100)), '-b', lw=1, alpha=1)
```
